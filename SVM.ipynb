# Implementation of ML Algorithms in Python 3.6.5

## 1. SVM

```python
from sklearn.svm import SVC 
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
 plt.subplot(1, 5, index + 1)
 plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
 plt.title('Training: %i\n' % label, fontsize = 20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=0)
```

```python
svclassifier = SVC(kernel='linear')  
svclassifier.fit(X_train, y_train)  
```

```python
svclassifier_new = SVC(kernel='poly',degree=3)
svclassifier_new.fit(X_train, y_train)  
```

```python
y_pred = svclassifier.predict(X_test)  
```

```python
y_pred_new=svclassifier_new.predict(X_test)
```

```python
# Use score method to get accuracy of model
score = svclassifier.score(X_test, y_test)

score_new=svclassifier_new.score(X_test, y_test)
print(score)
print(score_new)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
cm = metrics.confusion_matrix(y_test, y_pred)
print(cm)

cm_new = metrics.confusion_matrix(y_test, y_pred_new)
print(cm_new)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm_new, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```
